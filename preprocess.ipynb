```python
import os
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

# Path to the folder containing CSV files
folder_path = "C:/Users/ns-dhamne/Data Acquisition/ProcessedData/Normal"

# Uncomment for each folder and run again 
# folder_path = "C:/Users/ns-dhamne/Data Acquisition/ProcessedData/InnerRace"
# folder_path = "C:/Users/ns-dhamne/Data Acquisition/ProcessedData/OuterRace"

# Path to the new CSV file
output_file = "C:/Users/ns-dhamne/Data Acquisition/ProcessedData/Outerrace/n_summary_stats.csv"

# Uncomment for each folder and run again
output_file = "C:/Users/ns-dhamne/Data Acquisition/ProcessedData/Outerrace/ir_summary_stats.csv"
output_file = "C:/Users/ns-dhamne/Data Acquisition/ProcessedData/Outerrace/or_summary_stats.csv"

# List all CSV files in the folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Define a function to calculate statistics
def calculate_statistics(data):
    maximum = np.max(data)
    minimum = np.min(data)
    average = np.mean(data)
    std_dev = np.std(data)
    rms = np.sqrt(np.mean(data ** 2))
    skewness = skew(data)
    kurt = kurtosis(data)
    crest = maximum / rms
    form = rms / average
    return [maximum, minimum, average, std_dev, rms, skewness, kurt, crest, form]

# Create an empty DataFrame to store results
summary_df = pd.DataFrame(columns=['Maximum', 'Minimum', 'Average', 'Standard Deviation', 
                                   'Root Mean Square', 'Skewness', 'Kurtosis', 'Crest', 'Form', 'Condition'])

# Loop through each CSV file
for file_name in csv_files:
    file_path = os.path.join(folder_path, file_name)
    
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path, header=None)
    
    # Remove NaN values
    df.dropna(inplace=True)
    
    # Calculate number of batches
    num_batches = len(df) // 480
    
    # Initialize an empty list to store batch statistics
    batch_stats = []
    
    # Loop through each batch
    for i in range(num_batches):
        start_idx = i * 480
        end_idx = (i + 1) * 480
        
        # Extract batch
        batch_data = df.iloc[start_idx:end_idx, :].values.flatten()
        
        # Calculate statistics
        stats = calculate_statistics(batch_data)
        
        # Append statistics to batch list
        batch_stats.append({'Maximum': stats[0], 'Minimum': stats[1], 
                            'Average': stats[2], 'Standard Deviation': stats[3], 
                            'Root Mean Square': stats[4], 'Skewness': stats[5], 
                            'Kurtosis': stats[6], 'Crest': stats[7], 'Form': stats[8],
                            'Condition': 'Outerrace'})
        # Uncomment for each folder and run again
        # 'Condition': 'Outerrace'
        # 'Condition': 'Outerrace'
    
    # Convert batch statistics to DataFrame
    batch_df = pd.DataFrame(batch_stats)
    
    # Concatenate batch DataFrame with summary DataFrame
    summary_df = pd.concat([summary_df, batch_df], ignore_index=True)

# Write the summary DataFrame to a new CSV file
summary_df.to_csv(output_file, index=False)
```
